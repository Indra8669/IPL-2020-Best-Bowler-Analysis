```python
import pandas as pd #importing the library
```

```python
deliveries = pd.read_csv('drive/MyDrive/all_matches.csv') #imported the dataset
```

```python
deliveries
```

```python
deliveries_2020 = deliveries[deliveries["season"] == "2020/21"] # filtered the dataset
```

```python
deliveries_2020
```

```python
unique_bowlers = deliveries_2020["bowler"].unique().tolist() # getting the list of all bowlers
```

```python
unique_bowlers
```

```python
# calculated and stored the number of balls bowled by each bowler
balls_list = []
for bowler in unique_bowlers:
  bowler_df = deliveries_2020[deliveries_2020["bowler"] == bowler]
  balls = len(bowler_df)
  balls_list.append([bowler, balls])
```

```python
balls_list
```

```python
balls_df = pd.DataFrame(balls_list, columns = ["player", "balls"]) # converted the list to dataframe for ease in filtering out
```

```python
balls_df
```

```python
balls_df = balls_df[balls_df["balls"] >= 200] #filteredd the dataset to get bowler bowled more than 200 deliveries
```

```python
balls_df
```

```python
all_bowlers = balls_df["player"].tolist() #getting the list of bowlers satisfying our criteria
```

```python
all_bowlers
```

```python
# calculated avg, sr and eco of a bowler
data_bowlers = []
for bowler in all_bowlers:
  bowler_df = deliveries_2020[deliveries_2020["bowler"] == bowler]
  balls = len(bowler_df[(pd.isna(bowler_df["wides"])) & (pd.isna(bowler_df["noballs"]))])
  runs_bat = bowler_df[(pd.isna(bowler_df["byes"])) & (pd.isna(bowler_df["legbyes"]))]["runs_off_bat"].sum()
  runs_extras = bowler_df[(pd.isna(bowler_df["byes"])) & (pd.isna(bowler_df["legbyes"]))]["extras"].sum()
  total_runs = runs_bat + runs_extras
  wickets = len(bowler_df[pd.isna(bowler_df["player_dismissed"]) == False])
  average = total_runs/wickets
  econ = (total_runs*6)/balls
  sr = balls/wickets
  data_bowlers.append([bowler, average, econ, sr])
```

```python
data_bowlers
```

```python
# calculated the index for each bowler (sum of reciprocals of avg, sr and eco)
index_vals = []
for bowler_data in data_bowlers:
  index = (1/bowler_data[1]) + (1/bowler_data[2]) + (1/bowler_data[3])
  index_vals.append([bowler_data[0], index])
```

```python
index_vals
```

```python
index_df = pd.DataFrame(index_vals, columns = ["player", "index"]) #converted the list to dataframe, for sorting the data in descending order
```

```python
index_df
```
